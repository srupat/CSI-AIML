# Decision Tree for Iris Flower Classification 

#### Setting up the Libraries

```python
from sklearn.datasets import load_iris
import pandas as pd
```

#### Loading the dataset

```python
iris = load_iris()
print(iris)
```

```python
print(iris.DESCR)
```

#### Pandas Dataframe

```python
df = pd.DataFrame(iris['data'],columns=['sepal length','sepal width','petal length','petal width'])
print(df)
```

```python
type = list()
for i in range(50):
    type.append(0)
for i in range(50):
    type.append(1)
for i in range(50):
    type.append(2)  
print(type)  
print(len(type))    
```

#### Appending type column

```python
df['type'] = type
print(df)
```

```python
inputs = df.drop('type',axis='columns')
target = df['type']
print(type)
```

#### Import decision tree classifier model

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

#### Test

```python
model.predict([[5.0,3.6,1.4,0.2]])
```

```python
model.predict([[6.7,3.0,5.2,2.3]])
```

#### Visualization

```python
from matplotlib import pyplot as plt
```

```python
clf = tree.DecisionTreeClassifier(random_state=0)
clf.fit(inputs,target)
```

```python
tree.plot_tree(clf)
```

### Statistics 

We can use the describe command to get summary statistics of an array of data

```python
import scipy.stats as st
sum_stats = st.describe(iris.data)
print(sum_stats.minmax) #print min and max of sepal length, sepal width, petal length, petal width 
```

```python
print(sum_stats.mean) #print mean of sl,sw,pl,pw
print(sum_stats.variance)
```

```python
#Class 0
sum_stats = st.describe(iris.data[:50,:])
print(sum_stats.minmax)
print(sum_stats.mean)
print(sum_stats.variance)
```

### Trying out histograms for Class 0 of the dataset

X-axis : Values of the particular feature<br>
Y-axis : Distribution

```python
plt.subplot(221)
plt.hist(iris.data[:50,0])
plt.xlim([0,6])
plt.title('SL')
plt.subplot(222)
plt.hist(iris.data[:50,1])
plt.xlim([0,6])
plt.title('SW')
plt.subplot(223)
plt.hist(iris.data[:50,2])
plt.xlim([0,6])
plt.title('PL')
plt.subplot(224)
plt.hist(iris.data[:50,3])
plt.xlim([0,6])
plt.title('PW')
plt.suptitle('Class 0') #super title
plt.tight_layout() #to ensure proper arrangement and spacing
plt.show()
```

```python
import numpy as np

fig = plt.figure(figsize=(10,5), dpi = 150) #dots per inch
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

H, _x, _y = np.histogram2d(iris.data[:50,0], iris.data[:50,1]) #sepal length and sepal width
_xx, _yy = np.meshgrid(_x[1:], _y[1:]) #creates a grid for x and y values based on bin edges
xx, yy, HH = _xx.ravel(), _yy.ravel(), H.ravel() #ravel is used to convert a multidimensional array to a single array 

ax1.bar3d(xx, yy, np.zeros_like(HH), 0.13, 0.13, HH, shade=True, color='b') 

ax1.set_xlabel('SL')
ax1.set_ylabel('SW')

H, _x, _y = np.histogram2d(iris.data[:50,2], iris.data[:50,3])
_xx, _yy = np.meshgrid(_x[1:], _y[1:])
xx, yy, HH = _xx.ravel(), _yy.ravel(), H.ravel()

ax2.bar3d(xx, yy, np.zeros_like(HH), 0.04, 0.04, HH, shade=True, color='g')

ax2.set_xlabel('PL')
ax2.set_ylabel('PW')
```
